## Annotations storing

Coco way to store annotations
```json
{
      "id": 0,
      "iscrowd": 0,
      "image_id": 1,
      "category_id": 2,
      "segmentation": [
        [
          1611.9305019305014, 638.7001287001284, 1657.2458172458166,
          625.7528957528955, 1647.535392535392, 577.200772200772,
          1582.7992277992273, 573.9639639639637
        ]
      ],
      "bbox": [
        1582.7992277992273, 573.9639639639637, 74.44658944658931,
        64.73616473616471
      ],
      "area": 3211.178297713036
}
```

Vgg way to store annotations
```json
{
      "0": {
        "shape_attributes": {
          "name": "polygon",
          "all_points_x": [
            1611.9305019305014, 1657.2458172458166, 1647.535392535392,
            1582.7992277992273, 1611.9305019305014
          ],
          "all_points_y": [
            638.7001287001284, 625.7528957528955, 577.200772200772,
            573.9639639639637, 638.7001287001284
          ]
        },
        "region_attributes": { "label": "Levure" }
      }
```



In the coco storing style,
```python
[
1611.9305019305014, 638.7001287001284, 1657.2458172458166,
625.7528957528955, 1647.535392535392, 577.200772200772,
1582.7992277992273, 573.9639639639637
]
```
corresponds to the list of vertices $(x_1, y_1, x_2, y_2, \dots, x_n, y_n)$ of the given polygon where $x_i, y_i$ correspond to the $x$ and $y$ coordinates of the $i$-th vertex, while `all_points_x` and `all_points_y` in the vgg style correspond to $[x_1, x_2, \dots, x_n, x_1]$ and $[y_1, y_2, \dots, y_n, y_1]$. Note the $x_1$ and $y_1$ at the end of the two lists.

````python
lst = [
    1611.9305019305014,
    638.7001287001284,
    1657.2458172458166,
    625.7528957528955,
    1647.535392535392,
    577.200772200772,
    1582.7992277992273,
    573.9639639639637,
]
````

````python
even = [idx for idx in range(len(lst)) if idx % 2 == 0]
odd = [idx for idx in range(len(lst)) if idx % 2 == 1]
````
